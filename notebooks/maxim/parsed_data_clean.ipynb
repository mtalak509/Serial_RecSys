```python
def director_filler(line):
    if pd.isna(line):
        line = 'Unknown'
    return line

def imageURL_filler(line):
    if pd.isna(line):
        line = 'https://cdn-user30887.skyeng.ru/uploads/67699855aef14490296321.webp'
    return line

def actors_filler(line):
    if pd.isna(line):
        line = 'Unknown'
    return line

def year_filler(line):
    if pd.isna(line):
        line = 2000
    return line
```

https://cdn-user30887.skyeng.ru/uploads/67699855aef14490296321.webp

```python
data['director'] = data['director'].apply(director_filler)
data['image_url'] = data['image_url'].apply(imageURL_filler)
data['actors'] = data['actors'].apply(actors_filler)
data['year'] = data['year'].apply(actors_filler)
```

```python
data.dropna(axis=0, inplace=True)
```

```python
data.head(2)
```

```python
def genre_clean(line):
    ban_list = ['Adult', 'Music', 'DIY', 'Travel', 'Medical', 'Espionage', 'Supernatural']
    merge_dict = {'Children': 'Kids',
                  'kids': 'Kids',
                  'War': 'War&Politics',
                  'Adventure': 'Action&Adventure'}
    line = line.replace(';', ',')
    line = line.replace(' ', '')
    line_list = [item for item in line.split(',') if item not in ban_list]
    line_list = [merge_dict.get(genre, genre) for genre in line_list]
    line_list = list(set(line_list))
    line = ','.join(line_list)
    return line
```

```python
data['genres'] = data['genres'].apply(genre_clean)
```

```python
genre_dict = {}
def genre_count(line):
    # line = line.replace(';', ',')
    # line = line.replace(' ', '')
    for key in line.split(','):
        genre_dict[key] = genre_dict.setdefault(key, 0) + 1
    return genre_dict
```

```python
data['genres'].apply(genre_count)
```

```python
genre_dict = dict(sorted(genre_dict.items(), key=lambda x: x[1], reverse=True))
genre_dict
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.bar(genre_dict.keys(), genre_dict.values())
plt.xticks(rotation=90)
plt.show()
```
